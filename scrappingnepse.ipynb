```python
import requests
from bs4 import BeautifulSoup

url = BeautifulSoup("http://www.nepalstock.com/main/todays_price/index/1",'html.parser')
soup= requests.get(url)
soup
```

```python
soup.text
```

```python
lxml_code=BeautifulSoup(soup.text,'lxml')
lxml_code
```

```python
table_code = lxml_code.table
table_code
```

```python
tags= table_code.find_all('tr')
tags
```

```python
for tag in tags:
    print(tag.text)
```

```python
data=[]
for tag in tags:
    data.append(tag.text.split('\n')[1:-1])
    
data
```

```python
data
```

```python
import csv
f=open('datas.csv','w')
x=csv.writer(f)
for i in data:
    x.writerow(i)
    
f.close()
```
